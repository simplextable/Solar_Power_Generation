```python
X = df_merge["DAILY_YIELD"].values
size = int(len(X) * 0.8)
train, val = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
```

```python
train
```

```python
for t in range(len(val)):
    model = ARIMA(history, order=(0,0,0))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = val[t]
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = df_merge[["AMBIENT_TEMPERATURE","MODULE_TEMPERATURE", "Irradiation_Sum"]]
y = df_merge["daily_yield"]
```

```python
x_train, x_val, y_train, y_val = train_test_split(X, y, test_size=0.9, random_state=42)
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor(n_estimators = 10, random_state=0)
rf_reg.fit(x_train,y_train)
y_pred = rf_reg.predict(x_val)
pred_df = pd.DataFrame(data = y_pred, columns=['predictions'])
pred_df["true_values"] = y_val.values
```

```python
pred_df
```

```python
pyplot.plot(pred_df.predictions)
pyplot.plot(pred_df.true_values, color='red')
pyplot.show()
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

lr = linear_model.LinearRegression()

# Train the model using the training sets
lr.fit(x_train, y_train)

# Make predictions using the testing set
y_pred = lr.predict(x_val)

# The coefficients
print("Coefficients: \n", lr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_val, y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_val, y_pred))
#
print("MAE:", mean_absolute_error(y_val, y_pred))
```
